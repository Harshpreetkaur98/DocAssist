```python
import pandas as pd

df = pd.read_csv('data.csv')

df.head()
```

```python
df.isnull().sum()
# checking for null values
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
# Filling in the misisng values of Gender, Region, Ethnicity, SocioEconomic status, Previous Visits and Insurance Status with the mode of the respective columns
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Region'] = df['Region'].fillna(df['Region'].mode()[0])
df['Ethnicity'] = df['Ethnicity'].fillna(df['Ethnicity'].mode()[0])
df['Socioeconomic Status'] = df['Socioeconomic Status'].fillna(df['Socioeconomic Status'].mode()[0])
df['Previous Visits'] = df['Previous Visits'].fillna(df['Previous Visits'].mode()[0])
df['Insurance Status'] = df['Insurance Status'].fillna(df['Insurance Status'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Dropping rest of the missing data
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# Converting all the data to lowercase
data_lowercase = ['Gender', 'Region', 'Ethnicity', 'Socioeconomic Status', 'Symptom Description', 'Additional Symptoms', 'Chronic Condition','Allergies', 'Potential Condition', 'Suggested Action', 'Insurance Status']

for column in data_lowercase:
    df[column] = df[column].str.lower()

df.head()
```

```python
#  Dropping any duplicate rows
df = df.drop_duplicates()
```

```python
# Correcting Invalid data
df = df[(df['Age'] >= 0) & (df['Age'] <= 120)]
df = df[df['Gender'].isin(['male', 'female'])]
```
